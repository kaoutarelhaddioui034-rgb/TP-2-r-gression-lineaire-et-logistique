````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc
import os

# Téléchargement et chargement des données
url = 'https://stats.idre.ucla.edu/stat/data/binary.csv'
df = pd.read_csv(url)

print("Aperçu des données :")
display(df.head())
````

```markdown
### 1. Préparation des données
Nous transformons la variable catégorielle `prestige` en variables indicatrices (dummies) pour le modèle.
```

````python
# Encodage de la colonne 'rank' (nom réel dans le dataset)
dummies = pd.get_dummies(df['rank'], prefix='prestige')
data_final = df[['admit', 'gre', 'gpa']].join(dummies)

# Définition de X et y
X = data_final.drop('admit', axis=1)
y = data_final['admit']

# Division Train/Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Taille de l'ensemble d'entraînement : {len(X_train)}")
print(f"Taille de l'ensemble de test : {len(X_test)}")
````

### 2. Entraînement du modèle de Régression Logistique

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement d'un dataset contenant la variable MPG (Miles Per Gallon)
url_mpg = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'Model Year', 'Origin']
df_mpg = pd.read_csv(url_mpg, names=column_names, na_values='?', comment='\t', sep=' ', skipinitialspace=True)

# Premier diagramme : Distribution de la variable MPG
plt.figure(figsize=(10, 6))
sns.histplot(df_mpg['MPG'], kde=True, color='teal')
plt.title('Distribution de la consommation (MPG)')
plt.xlabel('Miles Per Gallon (MPG)')
plt.ylabel('Fréquence')
plt.show()
````

### 2. Diagramme de Régression (MPG)
Visualisation de la relation linéaire entre la puissance (Horsepower) et la consommation (MPG).

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Création du diagramme de régression
plt.figure(figsize=(10, 6))
sns.regplot(x='Horsepower', y='MPG', data=df_mpg,
            scatter_kws={'alpha':0.5},
            line_kws={'color':'red'})

plt.title('Régression Linéaire : Puissance vs Consommation')
plt.xlabel('Puissance (Horsepower)')
plt.ylabel('Miles Per Gallon (MPG)')
plt.grid(True, alpha=0.3)
plt.show()
````

### 3. Matrice de Corrélation (Tableau Coloré)
Visualisation des corrélations entre toutes les variables numériques du dataset Auto MPG.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcul de la matrice de corrélation
corr_matrix = df_mpg.corr(numeric_only=True)

# Création de la heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Matrice de Corrélation des variables Auto MPG')
plt.show()
````
